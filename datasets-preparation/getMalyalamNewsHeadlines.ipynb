```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
geckodriver = '/home/gaurav/Downloads/geckodriver'
 
options = webdriver.FirefoxOptions()
options.add_argument('-headless')
```

```python
url = 'https://m.dailyhunt.in/news/india/malayalam/-topics-8'
```

```python
browser = webdriver.Firefox(executable_path=geckodriver, firefox_options=options)
browser.get(url)
for i in range(1,200):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(4)
    print(i)
    
html = browser.page_source
 
browser.quit()
```

```python
entertainment_headings = []
data = html.encode('utf-8')
soup = BeautifulSoup(data, 'html.parser')
articles = soup.find_all('li',{'class':'lang_ml'})
for article in articles:
    entertainment_headings.append(article.find('h2').text)
```
